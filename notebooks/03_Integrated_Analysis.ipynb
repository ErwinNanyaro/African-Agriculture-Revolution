### Comparing Morogoro (strawberry) and Arusha (multi-crop) farmers

**Key Metrics:**
- 298 total farmers analyzed
- 2 regions: Morogoro & Arusha
- $10,000-$50,000 revenue potential

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

print("‚úÖ Libraries imported successfully")
print("üåç African Agriculture Revolution - Integrated Analysis")
```

```python
# Load both datasets
strawberry = pd.read_csv("../data/raw/strawberry_farmers_morogoro.csv")
arusha = pd.read_csv("../data/raw/both_crops_farmers_arusha.csv")

print("üìä DATASET COMPARISON:")
print(f"üçì Morogoro Strawberry: {len(strawberry)} farmers")
print(f"üåΩ Arusha Multi-Crop: {len(arusha)} farmers")
print(f"üáπüáø TOTAL TANZANIAN FARMERS: {len(strawberry) + len(arusha)}")

# Create simple visualization
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(8, 5))
regions = ["Morogoro\nStrawberry", "Arusha\nMulti-Crop", "Total"]
counts = [len(strawberry), len(arusha), len(strawberry) + len(arusha)]
colors = ["#FF6B6B", "#06D6A0", "#118AB2"]

bars = ax.bar(regions, counts, color=colors, edgecolor="black")
ax.set_title("Tanzanian Farmers Analyzed", fontweight="bold", fontsize=14)
ax.set_ylabel("Number of Farmers", fontweight="bold")

# Add value labels on bars
for bar, count in zip(bars, counts):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2., height + 5,
            f"{count}", ha="center", va="bottom", fontweight="bold")

plt.tight_layout()
plt.show()
```
